```python
import numpy as np 
```

```python
#создаем массив, проверяем размер (количество элементов)
a = np.array ([[1, 6],
               [2, 8],
               [3, 11],
               [3, 10],
               [1, 7]])
print(a.size)
```

```python
#находим среднее по столбцам и записываем результат в новый массив
mean_a = a.mean(axis = 0)
print (mean_a)
```

```python
#вычитаем среднее по столбцам
a_centered = a - mean_a
print(a_centered)
```

```python
#транспонируем массив для дальнейшего создания векторов
a_centered_t=a_centered.T
a_centered_t
```

```python
#подготовливаем 2 вектора из полученного массива с помощью срезов для вычисления скалярного произведения
c = a_centered_t[0, :].copy()
b = a_centered_t[1:,:].copy()
print(c)
print(b)
```

```python
#вычисляем скалярное произведение
a_centered_sp_=np.inner(c, b)
print(a_centered_sp_)
```

```python
#ручная проверка: создаем векторы и вычисляем скалярное произведение
d = [-1, 0, 1, 1, -1]
f = [-2.4, -0.4, 2.6, 1.6, -1.4]
a_centered_sp = np.inner(d, f)
print(a_centered_sp)
```

```python
#находим ковариацию с помощью деления скалярное произведения на количество наблюдений-1 (N-1=4)
cov = a_centered_sp / 4
print(cov)
```

**cov- несмещенная оценка ковариации**

```python
#получаем  m -транспонированный массив a с помощью атрибута .Т и далее используем функцию cov для расчета ковариации (в качестве аргумента передаем m)
m = a.T
m_cov = np.cov(m)
print(m)
print(m_cov)
print(m_cov[0, 1])
```

### Работа с данными в Pandas

```python
import pandas as pd
```

```python
df_authors = pd.DataFrame({'author_id':[1, 2, 3],
                           'author_name':['Тургенев','Чехов', 'Островский']}, columns=['author_id', 'author_name'])
```

```python
df_authors
```

```python
df_book = pd.DataFrame({'author_id':[1, 1, 1, 2, 2, 3, 3],
                        'book_title':['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],
                        'price':[450, 300, 350, 500, 450, 370,290]}, columns=['author_id', 'book_title', 'price'])
```

```python
df_book
```

```python
#используем функцию merge для слияния 2-х подготовленных df - внешнее слияние (how='outer') с индексацией по автору 
df_authors_price = pd.merge(df_authors, df_book, on='author_id', how='outer')
```

```python
df_authors_price
```

```python
#создаем df top5 с выводом строк из author_price с 5-тью самыми дорогими книгами с помощью метода .nlargest
df_top5=df_authors_price.nlargest(5, 'price')
```

```python
df_top5
```

```python
# группируем данные для рассчета основных статистик по ценам книг для каждого автора, используя .groupy  и .agg
authors_stat=df_authors_price.groupby('author_name').agg({'price':['max', 'min', 'mean']})
```

```python
authors_stat
```

```python
#используем функцию map и метод .join для объединения строк. Переименовываем столбцы со статистиками, согласно заданию. 
#Сбрасываем индексацию и назначаем новую по автору
authors_stat.columns = authors_stat.columns.map('_'.join)
authors_stat = authors_stat.rename(columns={'max':'max_price', 'min':'min_price', 'mean':'mean_price'})
authors_stat = authors_stat.reset_index()
authors_stat.set_index('author_name', inplace=True)

#authors_stat.columns = authors_stat.columns.get_level_values(0)- не сработало почему-то???
authors_stat
```

##Задание 5. Работа с pd.pivot_table


```python
#создаем новый столбец cover и заполняем его  списком типов обложек
df_authors_price.insert(3, 'cover', ['твердая', 'мягкая', 'мягкая', 'твердая','твердая', 'мягкая', 'мягкая'], False)
df_authors_price
```

```python
#присваиваем новое имя файлу, в котором рассчитавыем суммарную стоимость книг в твердой и мягкой обложках по каждому автору, пропуски заполняем нулями
book_info = pd.pivot_table(data= df_authors_price,
               values='price',
               index='author_name',
               columns='cover',
               aggfunc='sum',
               fill_value=0)
book_info
```

```python
#сохраняем файл в формате pickle и читаем его, присваивая новое имя, проверяе идентичность файлов book_info и book_info2
book_info.to_pickle('book_info.pkl')
book_info2=pd.read_pickle('book_info.pkl')
book_info2

```
